##5. Set Up Great Expectations Context and Data Source

```python
import great_expectations as gx

# Create a GE context
context = gx.get_context()

# Load your DataFrame into GE
batch = context.sources.pandas_default.read_dataframe(df)

# Define 3 expectations
batch.expect_column_values_to_be_between("Car1_Location_X", min_value=-100, max_value=100)
batch.expect_column_values_to_be_in_set("Car2_Location_Y", [140, 141])
batch.expect_column_values_to_not_be_null("Occluded_Image_view")
```
